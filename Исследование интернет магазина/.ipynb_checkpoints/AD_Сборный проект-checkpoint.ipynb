Данный проект выполнен для оценки компании "Стримчик" на 2017 год. В том числе, поставлены следующие задачи:  
- Определить сколько игр выпускалось в разные годы.

## Предобработка данных

### 1. Приведем названия столбцов к нижнем регистру

```python
data.columns = data.columns.str.lower()
#приведем наименования столбцов к нижнему регистру

data.head()
#посмотрим, что все впорядке
```

```python
data['name'] = data['name'].str.lower()
data['platform'] = data['platform'].str.lower()
data['genre'] = data['genre'].str.lower()
data['rating'] = data['rating'].str.lower()
#приведем значения в столбцах к нижнему регистру

data.head()
#посмотрим, что все впорядке
```

### 2. Заменим значания `K-A` в столбце `rating` на `e`.

```python
data.loc[data.rating == 'k-a', 'rating'] = 'e'
```

### 3. Отсортируем данные по году выпуска в порядке убывания.

```python
data = data.sort_values(by='year_of_release')
```

### 4. Заменим значения `tbd` в столбце `user_score` на NaN и приведем столбце к типу `float`.

```python
data.loc[data.user_score == 'tbd', 'user_score'] =  np.nan
#заменим значания на NaN

data.user_score = data.user_score.astype('float')
#приведем к типу float

data.info()
#посмотрим, что все впорядке
```

### 5. Удалим строки с пропущенными данными из столбца `year_of_release` и изменим тип данных на `int`.

```python
data = data.dropna(subset='year_of_release')
#удалим строки с пропусками

data = data.reset_index(drop=True)
#обновим индексы

data.year_of_release = data.year_of_release.astype('int')
#приведем данные в столбце к типу int

data.info()
#посмотрим, что все впорядке
```

### 6. Удалим пропущенные значения в столбце `name`.

```python
data = data.dropna(subset='name')
#удалим значания

data = data.reset_index(drop=True)
#обновим индексы

data.info()
#посмотрим, что все впорядке
```

### 7. Добавим столбец с суммарными продажами для каждой игры в датафрейм.

```python
data['sum_of_sales'] = data.na_sales + data.eu_sales + data.jp_sales + data.other_sales
#добавим новым столбец с данными

data.head()
#посмотрим, что все впорядке
```

### 8. Вывод

В части предобработки данных были скорректированны данные внутри датафрейма, удалены повторяющиеся данные и строки с отсутствующими данными.  
Так же в датафрейм был добавлен новый столбец с суммарными продажами по играм.

## Исследовательский анализ данных

### 1. Изменение количества игр по годам выпуска.

```python
(
    data
    .pivot_table(index='year_of_release', values='name', aggfunc='count') #отфильтруем данные, посчитаем количество игр по названию
    .plot(grid=True, sharex=False, figsize=(12,5)) #нарисуем график
)

plt.title('Количество выпускаемых игр по годам')
plt.xlabel('Year', color='gray')
plt.ylabel('Quntity', color='gray')
plt.legend(['Количество игр'], loc=2)
plt.show()
#зададим параметры оглавление, подписи по осям, легенда
```

Как видно из графика, до 1990 года количество выпускаемых игр было очень малым из-за того, что компьютерная техника была несильно развита и в основном использовалась для промышленных применений. Далее количество игр начиная с 1990 годов возрастало до ~2007 года и затем наблюдается спад. Тенденция графика направлена вниз. Возможно, стоит ожидать продолжение спада в будущем.

### 2. Сравнение продаж по платформам.

```python
(
    data
    .pivot_table(index='platform', values='sum_of_sales', aggfunc='sum') #отфильтруем данные, посчитаем продажи по платформам
    .sort_values(by='sum_of_sales', ascending=False) #отсортируем данные по убыванию
    .plot(grid=True, kind='bar', sharex=False, figsize=(12,5)) #построим диаграмму
)

plt.title('Продажи по платформам')
plt.xlabel('Платформа', color='gray')
plt.ylabel('Продажи', color='gray')
plt.legend(['Суммарные продажи'])
plt.show()
#зададим параметры оглавление, подписи по осям, легенда
```

Как видно из графика, самые продаваемые игры выпускаются на платформах `PS2`, `X360`, `PS3`, `WII` и `DS`.  
Построим график по данным топ-5 самых успешных платформ по годам

```python
top_sales_platform = ['ps2','x360','ps3','wii','ds'] #создадим переменную с топ-5 платформ

(
    data
    .loc[data.platform.isin(top_sales_platform) == True, ['platform', 'year_of_release', 'sum_of_sales']] #отсортируем необходимые столбцы с необходимыми данными из всего датафрейма
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='sum') #отфильтруем данные и посчитаем суммарно продажи по году и платформе
    .reset_index() #обнулим индексы
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform') #развернем таблицу
    .plot(grid=True, sharex=False, figsize=(12,5), xlim=(2000, 2017), style='o-') #нарисуем график
)


plt.title('Продажи топ-5 платформ по годам')
plt.xlabel('Год', color='gray')
plt.ylabel('Продажи', color='gray')
plt.show()
#зададим параметры оглавление, подписи по осям, легенда
```

Как видно из графика, большинство платформ полностью теряют свою актуальность в течении приблизительно 10 лет.

### 3. Определение самых продаваемых платформ.

Для того, чтобы верно сформулировать прогноз на будущее возьмем срез данных с 2008 года по 2016.

```python
data_last_years = data.loc[(data.year_of_release >= 2008)] #возьмем срез данных за 2008-2016 года

(
    data_last_years #возьмем датафрейм за 2008-2016 года
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='sum') #отфильтруем данные и посчитаем суммарно продажи по году и платформе
    .reset_index() #обнулим индексы
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform') #развернем таблицу
    .plot(grid=True, sharex=False, figsize=(12,5), style='o-', xlim=(2007, 2018)) #нарисуем график
)

plt.title('Платформы по суммам продаж за 2008-2016 года')
plt.xlabel('Год', color='gray')
plt.ylabel('Продажи', color='gray')
plt.legend(loc=1)
plt.show()
#зададим параметры оглавление, подписи по осям, легенда
```

По графику видно, что все платформы теряют популярность. Самые популярные платформы по продажам в 2016 году это `PS4`, `XOne`, `3DS` и `PC`.

### 4. Изучение диаграммы размаха для суммарных продаж в разбивке по платформам.

Увидим, что данные в основном лежат в промежутке до единицы. По остальным параметрам данные имеют множественные выбросы.

## Проверка гипотез

### 1. Проверим гипотезу, что средние пользовательские рейтинги у платформ Xbox One и PC одинаковые. 

Сформулируем гипотезы:  
  
H0: средние пользовательские рейтинги у платформ `Xbox One` и `PC` равны \
H1: средние пользовательские рейтинги у платформ `Xbox One` и `PC` неравны.

Создадим 2 переменные и запишем туда срезы данных из датафрейма для платформ Xbox One и для PC для данных за 2008-2016 года.

```python
data_xb_last_years = data_last_years.loc[(data_last_years.platform == 'xone') & (data_last_years.user_score.isna() == False)] 
#запишем в новую переменную данные по платформе Xbox One и откинем значения NaN в столбце user_score
data_pc_last_years = data_last_years.loc[(data_last_years.platform == 'pc') & (data_last_years.user_score.isna() == False)]
#запишем в новую переменную данные по платформе PC и откинем значения NaN в столбце user_score
```

Проверим нормально ли распределены данные в этих двух переменных по столбцу `user_score` с помощью Q-Q plot.  
Для этого напишем функцию для отрисовка графика в числах Z-шкалы.

```python
def qq_plot(x, title): #создадим функцию для отрисовка графика qq-plot
    m = x.mean() #высчитаем медиану в выборке
    std = x.std() #высчитаем выборочное стандартное отклонение
    x = (x - m)/std #переведем данные в Z-шкалу

    sm.qqplot(x, line ='45') #создадим график Q-Q plot
    plt.title(title)
    plt.show()
```

```python
qq_plot(data_xb_last_years.user_score, 'График Q-Q plot для пользовательской оценки платформы Xbox One. 2008-2016') 
#построим график для данных из столбца user_score среза данных по платформе Xbox One
```

```python
qq_plot(data_pc_last_years.user_score, 'График Q-Q plot для пользовательской оценки платформы PC. 2008-2016')
#построим график для данных из столбца user_score среза данных по платформе PC
```

Так как данные в наших выборках не распределены нормально и нет возможности применить тест Стьюдента, то применим U-тест Манни-Уитни для сравнения двух средних.

```python
stat, p = st.stats.mannwhitneyu(data_xb_last_years.user_score, data_pc_last_years.user_score) #применим тест Манни-Уитни
print(f'stat={stat:.3f}, p={p:.3f}') #выведем значение статистики и p-value
if p > 0.05: #создадим условие, что при значении p-value больше статистической значимости в 5% не отклоняем H0 и если меньше, то отклоняем H0 и принимаем H1
    print('Нет досточных оснований, чтобы отклонить H0')
else:
    print('Есть достаточно оснований, чтобы отклонить H0, принимаем H1.')
```

С помощью теста мы можем откинуть H0 и приминаем гипотезу H1 о том, что среднии двух выборок неравны.

### 2. Проверим гипотезу, что средние пользовательские рейтинги у жанров Action и Sports разные.

Сформулируем гипотезы:  
  
H0: средние пользовательские рейтинги у жанров `Action` и `Sports` равны \
H1: средние пользовательские рейтинги у жанров `Action` и `Sports` неравны.

Создадим 2 переменные и запишем туда срезы данных из датафрейма для жанров Action и Sports за 2008-2016 года.

```python
data_action_last_years = data_last_years.loc[(data_last_years.genre == 'action') & (data_last_years.user_score.isna() == False)]
#запишем в новую переменную данные по жанру action и откинем значения NaN в столбце user_score
data_sports_last_years = data_last_years.loc[(data_last_years.genre == 'sports') & (data_last_years.user_score.isna() == False)]
#запишем в новую переменную данные по жанру sports и откинем значения NaN в столбце user_score
```

Проверим нормально ли распределены данные в этих двух переменных по столбцу `user_score` с помощью Q-Q plot.  

```python
qq_plot(data_action_last_years.user_score, 'График Q-Q plot для пользовательской оценки жанра Action. 2008-2016') #построим график для данных из столбца user_score среза данных по жанру Action
```

```python
qq_plot(data_sports_last_years.user_score, 'График Q-Q plot для пользовательской оценки жанра Sports. 2008-2016') #построим график для данных из столбца user_score среза данных по жанру Sports
```

Так как данные в наших выборках не распределены нормально и нет возможности применить тест Стьюдента, то применим U-тест Манни-Уитни для сравнения двух средних.

```python
stat, p = st.stats.mannwhitneyu(data_action_last_years.user_score, data_sports_last_years.user_score) #применим тест Манни-Уитни
print(f'stat={stat:.3f}, p={p:.3f}') #выведем значение статистики и p-value
if p > 0.05: #создадим условие, что при значении p-value больше статистической значимости в 5% не отклоняем H0 и если меньше, то отклоняем H0 и принимаем H1
    print('Нет досточных оснований, чтобы отклонить H0')
else:
    print('Есть достаточно оснований, чтобы отклонить H0, принимаем H1.')
```

Тест показал, что можем отклонить H0 и принимаем H1, т.е. среднии двух выборок неравны.

Из приведенных гипотез о равенстве средних ни одна не подтвердилась. В первой части мы проверяли равенство средних значений в выборках по платформам `Xbox One` и `PC`, а во второй равенство средних по выборкам в жанрах `Action` и `Sports`.

Выбор жанров (`role-playing`, `action`, `misc` и т.д.) скорее всего связан с культурными особенностями людей, проживающих в данном регионе. Стоит провести анализ отдельно для этого вопроса. В части рейтинга так же наибольшую популярность имеют рейтинг широкой категории `Е`, а на втором месте уже рейтинг `T`, т.е. для подростков. Возможно, тут возрст играющих ниже, чем в регионе `EU`. В основном, вывод по региону `JP` можно сделать такой, что в данном регионе требуется отдельный подход к продажам с учетом культурных особенностей.  
  
В первой части проверки гипотез мы проверяли равенство средних значений в выборках по платформам `Xbox One` и `PC`, а во второй равенство средних по выборкам в жанрах `Action` и `Sports`. Из приведенных гипотез о равенстве средних ни одна не подтвердилась. Это говорит о том, что данные жанры и платформы имеют разную популярность, что можно увидеть из графиков по продажам.  
  
В дальнейшим, рекомендую сделать больший упор на собираемости данных по рейтингам как от критиков так и от пользователей.
