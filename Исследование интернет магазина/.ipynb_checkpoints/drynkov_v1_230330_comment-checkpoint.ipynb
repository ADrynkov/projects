### 2. –ó–∞–º–µ–Ω–∏–º –∑–Ω–∞—á–∞–Ω–∏—è `K-A` –≤ —Å—Ç–æ–ª–±—Ü–µ `rating` –Ω–∞ `e`.

### 4. –ó–∞–º–µ–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è `tbd` –≤ —Å—Ç–æ–ª–±—Ü–µ `user_score` –Ω–∞ NaN –∏ –ø—Ä–∏–≤–µ–¥–µ–º —Å—Ç–æ–ª–±—Ü–µ –∫ —Ç–∏–ø—É `float`.

### 5. –£–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `year_of_release` –∏ –∏–∑–º–µ–Ω–∏–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ `int`.

```python
data = data.dropna(subset='year_of_release')
#—É–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏

data = data.reset_index(drop=True)
#–æ–±–Ω–æ–≤–∏–º –∏–Ω–¥–µ–∫—Å—ã

data.year_of_release = data.year_of_release.astype('int')
#–ø—Ä–∏–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ –∫ —Ç–∏–ø—É int

data.info()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

<div class="alert alert-danger">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: ‚ùå</b>

–ó–¥–µ—Å—å –≤—Å–µ —Ç–∞–∫–∏ –ø–æ—à–ª–æ —á—Ç–æ-—Ç–æ –Ω–µ —Ç–∞–∫. –ü–æ—Å–º–æ—Ç—Ä–∏ –ø–æ–∂–∞–ª—É–π—Å—Ç–∞ –≤ —á–µ–º –¥–µ–ª–æ –∑–¥–µ—Å—å –Ω–∞ —Å–µ—Ä–≤–µ—Ä–µ.
</div>

### 6. –£–¥–∞–ª–∏–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ `name`.

```python
data = data.dropna(subset='name')
#—É–¥–∞–ª–∏–º –∑–Ω–∞—á–∞–Ω–∏—è

data = data.reset_index(drop=True)
#–æ–±–Ω–æ–≤–∏–º –∏–Ω–¥–µ–∫—Å—ã

data.info()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

### 7. –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –∏–≥—Ä—ã –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.

```python
data['sum_of_sales'] = data.na_sales + data.eu_sales + data.jp_sales + data.other_sales
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–º —Å—Ç–æ–ª–±–µ—Ü —Å –¥–∞–Ω–Ω—ã–º–∏

data.head()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

### 8. –í—ã–≤–æ–¥

–í —á–∞—Å—Ç–∏ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –±—ã–ª–∏ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤–Ω—É—Ç—Ä–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, —É–¥–∞–ª–µ–Ω—ã –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –¥–∞–Ω–Ω—ã–µ –∏ —Å—Ç—Ä–æ–∫–∏ —Å –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏.  
–¢–∞–∫ –∂–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –±—ã–ª –¥–æ–±–∞–≤–ª–µ–Ω –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –ø–æ –∏–≥—Ä–∞–º.

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### 1. –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º –≤—ã–ø—É—Å–∫–∞.

```python
(
    data
    .pivot_table(index='year_of_release', values='name', aggfunc='count') #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ, –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é
    .plot(grid=True, sharex=False, figsize=(12,5)) #–Ω–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫
)

plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—Å–∫–∞–µ–º—ã—Ö –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º')
plt.xlabel('Year', color='gray')
plt.ylabel('Quntity', color='gray')
plt.legend(['–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä'], loc=2)
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –≥—Ä–∞—Ñ–∏–∫–∞, –¥–æ 1990 –≥–æ–¥–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—Å–∫–∞–µ–º—ã—Ö –∏–≥—Ä –±—ã–ª–æ –æ—á–µ–Ω—å –º–∞–ª—ã–º –∏–∑-–∑–∞ —Ç–æ–≥–æ, —á—Ç–æ –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–∞—è —Ç–µ—Ö–Ω–∏–∫–∞ –±—ã–ª–∞ –Ω–µ—Å–∏–ª—å–Ω–æ —Ä–∞–∑–≤–∏—Ç–∞ –∏ –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∞—Å—å –¥–ª—è –ø—Ä–æ–º—ã—à–ª–µ–Ω–Ω—ã—Ö –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–π. –î–∞–ª–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –Ω–∞—á–∏–Ω–∞—è —Å 1990 –≥–æ–¥–æ–≤ –≤–æ–∑—Ä–∞—Å—Ç–∞–ª–æ –¥–æ ~2007 –≥–æ–¥–∞ –∏ –∑–∞—Ç–µ–º –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è —Å–ø–∞–¥. –¢–µ–Ω–¥–µ–Ω—Ü–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∞ –≤–Ω–∏–∑. –í–æ–∑–º–æ–∂–Ω–æ, —Å—Ç–æ–∏—Ç –æ–∂–∏–¥–∞—Ç—å –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ —Å–ø–∞–¥–∞ –≤ –±—É–¥—É—â–µ–º.

### 2. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.

```python
(
    data
    .pivot_table(index='platform', values='sum_of_sales', aggfunc='sum') #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ, –ø–æ—Å—á–∏—Ç–∞–µ–º –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º
    .sort_values(by='sum_of_sales', ascending=False) #–æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —É–±—ã–≤–∞–Ω–∏—é
    .plot(grid=True, kind='bar', sharex=False, figsize=(12,5)) #–ø–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É
)

plt.title('–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º')
plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞', color='gray')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏', color='gray')
plt.legend(['–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏'])
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –≥—Ä–∞—Ñ–∏–∫–∞, —Å–∞–º—ã–µ –ø—Ä–æ–¥–∞–≤–∞–µ–º—ã–µ –∏–≥—Ä—ã –≤—ã–ø—É—Å–∫–∞—é—Ç—Å—è –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö `PS2`, `X360`, `PS3`, `WII` –∏ `DS`.  
–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–æ –¥–∞–Ω–Ω—ã–º —Ç–æ–ø-5 —Å–∞–º—ã—Ö —É—Å–ø–µ—à–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º

```python
top_sales_platform = ['ps2','x360','ps3','wii','ds'] #—Å–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é —Å —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º

(
    data
    .loc[data.platform.isin(top_sales_platform) == True, ['platform', 'year_of_release', 'sum_of_sales']] #–æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='sum') #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω–æ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ–¥—É –∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ
    .reset_index() #–æ–±–Ω—É–ª–∏–º –∏–Ω–¥–µ–∫—Å—ã
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform') #—Ä–∞–∑–≤–µ—Ä–Ω–µ–º —Ç–∞–±–ª–∏—Ü—É
    .plot(grid=True, sharex=False, figsize=(12,5), xlim=(2000, 2017), style='o-') #–Ω–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫
)


plt.title('–ü—Ä–æ–¥–∞–∂–∏ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º')
plt.xlabel('–ì–æ–¥', color='gray')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏', color='gray')
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –≥—Ä–∞—Ñ–∏–∫–∞, –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–ª–Ω–æ—Å—Ç—å—é —Ç–µ—Ä—è—é—Ç —Å–≤–æ—é –∞–∫—Ç—É–∞–ª—å–Ω–æ—Å—Ç—å –≤ —Ç–µ—á–µ–Ω–∏–∏ –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ 10 –ª–µ—Ç.

### 3. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∞–º—ã—Ö –ø—Ä–æ–¥–∞–≤–∞–µ–º—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º.

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –≤–µ—Ä–Ω–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ –±—É–¥—É—â–µ–µ –≤–æ–∑—å–º–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö —Å 2008 –≥–æ–¥–∞ –ø–æ 2016.

```python
data_last_years = data.loc[(data.year_of_release >= 2008)] #–≤–æ–∑—å–º–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –∑–∞ 2008-2016 –≥–æ–¥–∞

(
    data_last_years #–≤–æ–∑—å–º–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑–∞ 2008-2016 –≥–æ–¥–∞
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='sum') #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω–æ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ–¥—É –∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ
    .reset_index() #–æ–±–Ω—É–ª–∏–º –∏–Ω–¥–µ–∫—Å—ã
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform') #—Ä–∞–∑–≤–µ—Ä–Ω–µ–º —Ç–∞–±–ª–∏—Ü—É
    .plot(grid=True, sharex=False, figsize=(12,5), style='o-', xlim=(2007, 2018)) #–Ω–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫
)

plt.title('–ü–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ —Å—É–º–º–∞–º –ø—Ä–æ–¥–∞–∂ –∑–∞ 2008-2016 –≥–æ–¥–∞')
plt.xlabel('–ì–æ–¥', color='gray')
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏', color='gray')
plt.legend(loc=1)
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏–∑ —Å–∞–º—ã—Ö –ø—Ä–æ–¥–∞–≤–∞–µ–º—ã—Ö –∏ –æ—Ü–µ–Ω–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–í—ã–≤–æ–¥, —á—Ç–æ –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–µ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä. –û—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏–º–µ—é—Ç –≤–ª–∏—è–Ω–∏–µ –ø–æ—á—Ç–∏ –≤–æ –≤—Å–µ—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö, –Ω–æ –Ω–µ —Å–∏–ª—å–Ω–æ–µ.

### 6. –ò–∑—É—á–∏–º –ø—Ä–∏–±—ã–ª—å–Ω–æ—Å—Ç—å –ø–æ –∂–∞–Ω—Ä–∞–º.

–í —Ä–µ–≥–∏–æ–Ω–µ `JP` –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞–µ–º—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∏ –∂–∞–Ω—Ä—ã –∏–∑ —Ç–æ–ø-5 –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤ –∏ —Å –æ–±—â–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º. –¢–∞–∫ –∂–µ —Ç—É—Ç —è–≤–Ω–æ –ø—Ä–µ–æ–±–ª–∞–¥–∞—é—Ç –∏–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã (`DS`, `3DS`, `PS3` –∏ —Ç.–¥.), –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω—ã –≤ —Ä–µ–≥–∏–æ–Ω–∞—Ö `EU` –∏ `NA`. –í–∏–¥–∏–º–æ, –Ω–∞ —Ç–µ—Ä—Ä–∏—Ç–æ—Ä–∏–∏ –¥–∞–Ω–Ω–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–Ω–æ–∂–µ—Å—Ç–≤–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Ç—Å—Ç–≤ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ –æ–Ω–∏ —Ç—É—Ç –±–æ–ª–µ–µ —à–∏—Ä–æ–∫–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã. –û–¥–Ω–∞–∫–æ, –º–æ–∂–Ω–æ –æ–±—Ä–∞—Ç–∏—Ç—å —Ç–∞–∫ –∂–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ –¥–∞–Ω–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ —Ç–∞–∫ –∂–µ —à–∏—Ä–æ–∫–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –∏–≥—Ä—ã –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ `PS2`, –∫–æ—Ç–æ—Ä–∞—è —É–∂–µ —É—Å—Ç–∞—Ä–µ–ª–∞. –í—ã–±–æ—Ä –∂–∞–Ω—Ä–æ–≤ (`role-playing`, `action`, `misc` –∏ —Ç.–¥.) —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ —Å–≤—è–∑–∞–Ω —Å –∫—É–ª—å—Ç—É—Ä–Ω—ã–º–∏ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—è–º–∏ –ª—é–¥–µ–π, –ø—Ä–æ–∂–∏–≤–∞—é—â–∏—Ö –≤ –¥–∞–Ω–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ. –°—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–Ω–∞–ª–∏–∑ –æ—Ç–¥–µ–ª—å–Ω–æ –¥–ª—è —ç—Ç–æ–≥–æ –≤–æ–ø—Ä–æ—Å–∞. –í —á–∞—Å—Ç–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞ —Ç–∞–∫ –∂–µ –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –∏–º–µ—é—Ç —Ä–µ–π—Ç–∏–Ω–≥ —à–∏—Ä–æ–∫–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ `–ï`, –∞ –Ω–∞ –≤—Ç–æ—Ä–æ–º –º–µ—Å—Ç–µ —É–∂–µ —Ä–µ–π—Ç–∏–Ω–≥ `T`, —Ç.–µ. –¥–ª—è –ø–æ–¥—Ä–æ—Å—Ç–∫–æ–≤. –í–æ–∑–º–æ–∂–Ω–æ, —Ç—É—Ç –≤–æ–∑—Ä—Å—Ç –∏–≥—Ä–∞—é—â–∏—Ö –Ω–∏–∂–µ, —á–µ–º –≤ —Ä–µ–≥–∏–æ–Ω–µ `EU`. –í –æ—Å–Ω–æ–≤–Ω–æ–º, –≤—ã–≤–æ–¥ –ø–æ —Ä–µ–≥–∏–æ–Ω—É `JP` –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫–æ–π, —á—Ç–æ –≤ –¥–∞–Ω–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è –æ—Ç–¥–µ–ª—å–Ω—ã–π –ø–æ–¥—Ö–æ–¥ –∫ –ø—Ä–æ–¥–∞–∂–∞–º —Å —É—á–µ—Ç–æ–º –∫—É–ª—å—Ç—É—Ä–Ω—ã—Ö –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–µ–π.

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

### 1. –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. 

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã:  
  
H0: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –ø–ª–∞—Ç—Ñ–æ—Ä–º `Xbox One` –∏ `PC` —Ä–∞–≤–Ω—ã \
H1: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –ø–ª–∞—Ç—Ñ–æ—Ä–º `Xbox One` –∏ `PC` –Ω–µ—Ä–∞–≤–Ω—ã.

–°–æ–∑–¥–∞–¥–∏–º 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ –∑–∞–ø–∏—à–µ–º —Ç—É–¥–∞ —Å—Ä–µ–∑—ã –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ –¥–ª—è PC –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∑–∞ 2008-2016 –≥–æ–¥–∞.

```python
data_xb_last_years = data_last_years.loc[(data_last_years.platform == 'xone') & (data_last_years.user_score.isna() == False)] 
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ Xbox One –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
data_pc_last_years = data_last_years.loc[(data_last_years.platform == 'pc') & (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PC –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
```

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤ —ç—Ç–∏—Ö –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É `user_score` —Å –ø–æ–º–æ—â—å—é Q-Q plot.  
–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞ –≤ —á–∏—Å–ª–∞—Ö Z-—à–∫–∞–ª—ã.

```python
def qq_plot(x, title): #—Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞ qq-plot
    m = x.mean() #–≤—ã—Å—á–∏—Ç–∞–µ–º –º–µ–¥–∏–∞–Ω—É –≤ –≤—ã–±–æ—Ä–∫–µ
    std = x.std() #–≤—ã—Å—á–∏—Ç–∞–µ–º –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
    x = (x - m)/std #–ø–µ—Ä–µ–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –≤ Z-—à–∫–∞–ª—É

    sm.qqplot(x, line ='45') #—Å–æ–∑–¥–∞–¥–∏–º –≥—Ä–∞—Ñ–∏–∫ Q-Q plot
    plt.title(title)
    plt.show()
```

```python
qq_plot(data_xb_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã Xbox One. 2008-2016') 
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ Xbox One
```

```python
qq_plot(data_pc_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã PC. 2008-2016')
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PC
```

–¢–∞–∫ –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ –≤ –Ω–∞—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏ –Ω–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–æ –ø—Ä–∏–º–µ–Ω–∏–º U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö.

```python
stat, p = st.stats.mannwhitneyu(data_xb_last_years.user_score, data_pc_last_years.user_score) #–ø—Ä–∏–º–µ–Ω–∏–º —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏
print(f'stat={stat:.3f}, p={p:.3f}') #–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value
if p > 0.05: #—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5% –Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1
    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

–° –ø–æ–º–æ—â—å—é —Ç–µ—Å—Ç–∞ –º—ã –º–æ–∂–µ–º –æ—Ç–∫–∏–Ω—É—Ç—å H0 –∏ –ø—Ä–∏–º–∏–Ω–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É H1 –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã.

### 2. –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã:  
  
H0: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ `Action` –∏ `Sports` —Ä–∞–≤–Ω—ã \
H1: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ `Action` –∏ `Sports` –Ω–µ—Ä–∞–≤–Ω—ã.

–°–æ–∑–¥–∞–¥–∏–º 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ –∑–∞–ø–∏—à–µ–º —Ç—É–¥–∞ —Å—Ä–µ–∑—ã –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –∑–∞ 2008-2016 –≥–æ–¥–∞.

```python
data_action_last_years = data_last_years.loc[(data_last_years.genre == 'action') & (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –∂–∞–Ω—Ä—É action –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
data_sports_last_years = data_last_years.loc[(data_last_years.genre == 'sports') & (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –∂–∞–Ω—Ä—É sports –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
```

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤ —ç—Ç–∏—Ö –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É `user_score` —Å –ø–æ–º–æ—â—å—é Q-Q plot.  

```python
qq_plot(data_action_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–∞ Action. 2008-2016') #–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä—É Action
```

```python
qq_plot(data_sports_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–∞ Sports. 2008-2016') #–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä—É Sports
```

–¢–∞–∫ –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ –≤ –Ω–∞—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏ –Ω–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–æ –ø—Ä–∏–º–µ–Ω–∏–º U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö.

```python
stat, p = st.stats.mannwhitneyu(data_action_last_years.user_score, data_sports_last_years.user_score) #–ø—Ä–∏–º–µ–Ω–∏–º —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏
print(f'stat={stat:.3f}, p={p:.3f}') #–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value
if p > 0.05: #—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5% –Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1
    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

–¢–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª, —á—Ç–æ –º–æ–∂–µ–º –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1, —Ç.–µ. —Å—Ä–µ–¥–Ω–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã.

### 3. –í—ã–≤–æ–¥

–ò–∑ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã—Ö –≥–∏–ø–æ—Ç–µ–∑ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –Ω–∏ –æ–¥–Ω–∞ –Ω–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–ª–∞—Å—å. –í –ø–µ—Ä–≤–æ–π —á–∞—Å—Ç–∏ –º—ã –ø—Ä–æ–≤–µ—Ä—è–ª–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –≤—ã–±–æ—Ä–∫–∞—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º `Xbox One` –∏ `PC`, –∞ –≤–æ –≤—Ç–æ—Ä–æ–π —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ –≤—ã–±–æ—Ä–∫–∞–º –≤ –∂–∞–Ω—Ä–∞—Ö `Action` –∏ `Sports`.

## –û–±—â–∏–π –≤—ã–≤–æ–¥

–î–∞–Ω–Ω—ã–π –ø—Ä–æ–µ–∫—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–æ–º–ø–∞–Ω–∏–∏ "–°—Ç—Ä–∏–º—á–∏–∫" –Ω–∞ 2017 –≥–æ–¥.  
–ü—Ä–∏ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–µ —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –±—ã–ª–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –æ—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã –≤ –≤–∏–¥–µ –º–Ω–æ–≥–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –Ω–µ–≤–µ—Ä–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ –Ω–µ–≤–µ—Ä–Ω—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö. –î–ª—è —É—Å—Ç—Ä–∞–Ω–µ–Ω–∏—è —ç—Ç–∏—Ö –ø—Ä–æ–±–ª–µ–º –±—ã–ª–∞ –ø—Ä–æ–≤–µ–¥–µ–Ω–∞ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö. –¢–∞–∫ –∂–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –±—ã–ª –¥–æ–±–∞–≤–ª–µ–Ω –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –ø–æ –∏–≥—Ä–∞–º –ø–æ –≤—Å–µ–º —Ä–µ–≥–∏–æ–Ω–∞–º.  
–ò–∑ –∏–∑—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —Ç–µ–Ω–¥–µ–Ω—Ü–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ –∂–∞–Ω—Ä–∞–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∞ –≤–Ω–∏–∑. –°—Ç–æ–∏—Ç –æ–∂–∏–¥–∞—Ç—å –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ —Å–ø–∞–¥–∞ –≤ –±—É–¥—É—â–µ–º –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ –≤—Å–µ—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –∏ –≤—Å–µ—Ö –∂–∞–Ω—Ä–æ–≤. 
–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ –≥–ª–æ–±–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å–∞–º—ã–º–∏ —É—Å–ø–µ—à–Ω—ã–º–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º–∏ –≤ 2016 –≥–æ–¥—É —è–≤–ª—è—é—Ç—Å—è `PS4`, `XOne`, `3DS` –∏ `PC`.  
–ï—Å–ª–∏ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å—Å—è –Ω–∞ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä, —Ç–æ —Å—Ç–æ–∏—Ç –±–æ–ª—å—à–µ –≤–Ω–∏–º–∞–Ω–∏—è —É–¥–µ–ª—è—Ç—å –æ—Ü–µ–Ω–∫–µ –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–µ–∂–µ–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. –û—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏–º–µ—é—Ç —è–≤–Ω–æ –±–û–ª—å—à–µ–µ –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏, —á–µ–º –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.  
–¢–∞–∫ –∂–µ —Å—Ç–æ–∏—Ç —É—á–µ—Å—Ç—å, —á—Ç–æ —Å–∞–º—ã–º–∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã–º–∏ –∏–≥—Ä–∞–º–∏ –≤ 2016-–æ–º –≥–æ–¥—É —è–≤–ª—è—é—Ç—Å—è `action`, `shooter`, `sports` –∏ `role-playing`. –¢–∞–∫ –∫–∞–∫ `shooter` –∏ `action` —ç—Ç–æ –µ—Å–ª–∏ –Ω–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ç–æ –∫–∞–∫ –º–∏–Ω–∏–º—É–º —Ä–æ–¥—Å—Ç–≤–µ–Ω–Ω—ã–µ –∂–∞–Ω—Ä—ã, —Ç–æ —Å—Ç–æ–∏—Ç –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∏–≥—Ä—ã —Ç–∞–∫–æ–≥–æ –∂–∞–Ω—Ä–∞ –∏–º–µ—é—Ç –ø–æ–¥–∞–≤–ª—è—é—â–µ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –Ω–µ–∂–µ–ª–∏ –¥—Ä—É–≥–∏–µ –∂–∞–Ω—Ä—ã. –ù–∞ –∏—Ö –ø—Ä–æ–¥–∞–∂–∏ —Å—Ç–æ–∏—Ç —Å–¥–µ–ª–∞—Ç—å —É–ø–æ—Ä.
