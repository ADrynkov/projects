Entertainment Software Rating Board). –≠—Ç–∞ –∞—Å—Å–æ—Ü–∏–∞—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Ä–µ–π—Ç–∏–Ω–≥ –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã—Ö –∏–≥—Ä –∏ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç –∏–º –ø–æ–¥—Ö–æ–¥—è—â—É—é –≤–æ–∑—Ä–∞—Å—Ç–Ω—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é.

### 3. –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –≥–æ–¥—É –≤—ã–ø—É—Å–∫–∞ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è.

### 5. –£–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `year_of_release` –∏ –∏–∑–º–µ–Ω–∏–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ `int`.

```python
data = data.dropna(subset=['year_of_release'])
#—É–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏

data = data.reset_index(drop=True)
#–æ–±–Ω–æ–≤–∏–º –∏–Ω–¥–µ–∫—Å—ã

data.year_of_release = data.year_of_release.astype('int')
#–ø—Ä–∏–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ –∫ —Ç–∏–ø—É int

data.info()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

### 6. –£–¥–∞–ª–∏–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ `name`.

```python
data = data.dropna(subset=['name'])
#—É–¥–∞–ª–∏–º –∑–Ω–∞—á–∞–Ω–∏—è

data = data.reset_index(drop=True)
#–æ–±–Ω–æ–≤–∏–º –∏–Ω–¥–µ–∫—Å—ã

data.info()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

### 7. –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –∏–≥—Ä—ã –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.

```python
data['sum_of_sales'] = data[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–º —Å—Ç–æ–ª–±–µ—Ü —Å –¥–∞–Ω–Ω—ã–º–∏

data.head()
#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –≤—Å–µ –≤–ø–æ—Ä—è–¥–∫–µ
```

### 8. –í—ã–≤–æ–¥

–í —á–∞—Å—Ç–∏ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –±—ã–ª–∏ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤–Ω—É—Ç—Ä–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, —É–¥–∞–ª–µ–Ω—ã –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –¥–∞–Ω–Ω—ã–µ –∏ —Å—Ç—Ä–æ–∫–∏ —Å –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏.

### 1. –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º –≤—ã–ø—É—Å–∫–∞.

```python
(
    data
    .pivot_table(index='year_of_release', values='name', aggfunc='count') 
    #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ, –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é
    .plot(grid=True, sharex=False, figsize=(12,5)) 
    #–Ω–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫
)

plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—Å–∫–∞–µ–º—ã—Ö –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º')
plt.xlabel('Year', color='gray')
plt.ylabel('Quntity', color='gray')
plt.legend(['–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä'], loc=2)
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–¢–∞–∫ –±—ã—Å—Ç—Ä–µ–µ –ø–æ–Ω—è—Ç—å –æ —á–µ–º –∏–¥—ë—Ç —Ä–µ—á—å –Ω–∞ —Å–∞–º–æ–º –≥—Ä–∞—Ñ–∏–∫–µ.

–ó–¥–æ—Ä–æ–≤–æ, —á—Ç–æ –≥—Ä–∞—Ñ–∏–∫ —É–≤–µ–ª–∏—á–µ–Ω üëç, —Ç–∞–∫ –≥–æ—Ä–∞–∑–¥–æ –ª–µ–≥—á–µ –∏–∑—É—á–∞—Ç—å —ç—Ç—É –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.

```python
top_sales_platform_name = ['nes','gb','2600','snes','x360'] #—Å–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é —Å —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º

#—Å–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –≥–¥–µ –±—É–¥—É—Ç –ø—Ä–æ–¥–∞–∂–∏ —Ç–æ–ª—å–∫–æ –ø–æ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º
top_sales_platform_df = (
    data
    .loc[data.platform.isin(top_sales_platform_name) == True, ['platform', 'year_of_release', 'sum_of_sales']]
    #–æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='median') 
    #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω–æ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ–¥—É –∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform') 
    #—Ä–∞–∑–≤–µ—Ä–Ω–µ–º —Ç–∞–±–ª–∏—Ü—É
    .reset_index() 
    #–æ–±–Ω—É–ª–∏–º –∏–Ω–¥–µ–∫—Å—ã
)

#—Å–æ–∑–¥–∞–¥–∏–º –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∏
fig, axs = plt.subplots(len(top_sales_platform_name),
                        sharex='row',
                        sharey=True,
                        figsize=(15, 15))

#–∑–∞–¥–∞–¥–∏–º –æ–±—â–µ–µ –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã
fig.suptitle('–ü—Ä–æ–¥–∞–∂–∏ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º') 

#—Å–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–º–µ–Ω–Ω—É—é –¥–ª—è –≤—Å–µ—Ö –ª–µ—Ç –æ—Ç 1985 –¥–æ 2017 –≥–æ–¥–æ–≤ –∏ –∑–∞–ø–æ–ª–Ω–∏–º –µ–µ –≤ —Ü–∏–∫–ª–µ
labels_xticks = []
for i in range(33):
    labels_xticks.append(1985+i)

#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –≤—Å–µ—Ö —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ —Å—É–º–º–∞–º –ø—Ä–æ–¥–∞–∂
for i,ax in enumerate(axs):
    axs[i].plot(top_sales_platform_df['year_of_release'], top_sales_platform_df.iloc[:,i+1], 'o-')  
    axs[i].set_xlim((1985, 2017)) #–Ω–∞—Å—Ç—Ä–æ–∏–º –≤–∏–¥–∏–º–æ—Å—Ç—å (–æ—Ç–∫–∏–Ω–µ–º –ª–∏—à–Ω—é—é —á–∞—Å—Ç—å –¥–æ 1995 –≥–æ–¥–∞)
    axs[i].grid() #–¥–æ–±–∞–≤–∏–º —Å–µ—Ç–∫—É
    axs[i].set_ylabel('–ü—Ä–æ–¥–∞–∂–∏, –º–ª–Ω.', color='gray') #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å–∏ y
    axs[i].xaxis.set_major_locator(ticker.FixedLocator(labels_xticks)) #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å–∏ —Ö
    axs[i].title.set_text(top_sales_platform_df.columns[i+1]) #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
    axs[i].set_xticklabels(labels_xticks, rotation=45) #–ø–æ–≤–µ—Ä–Ω–µ–º —Ç–∏–∫–∏ –Ω–∞ 45 –≥—Ä–∞–¥—É—Å–æ–≤

plt.subplots_adjust(hspace=0.5) #–¥–æ–±–∞–≤–∏–º –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –º–µ–∂–¥—É –≥—Ä–∞—Ñ–∏–∫–∞–º–∏
plt.show() #–æ—Ç–æ–±—Ä–∞–∑–∏–º –≥—Ä–∞—Ñ–∏–∫–∏
```

–°–ø–∞—Å–∏–±–æ –∑–∞ –ø–æ–¥—Å–∫–∞–∑–∫—É. –†–µ—à–∏–ª –ø–æ–∏–∑—É—á–∞—Ç—å subplots –∏ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –≤—Å–µ –≤ —Ü–∏–∫–ª–µ —Å–¥–µ–ª–∞—Ç—å. –í–æ—Ç —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å. –ú–æ–∂–µ—Ç –µ—Å—Ç—å —Å–æ–≤–µ—Ç—ã –∫–∞–∫ –∏—Ö –µ—â–µ –±–æ–ª–µ–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω–µ–µ –∏ –∫—Ä–∞—Å–∏–≤–µ–µ —Å–¥–µ–ª–∞—Ç—å?</div>

–ú–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –±–µ–∑ `subplots`, –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã `heatmap`. –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ —Å–¥–µ–ª–∞–≤ —Ç–∞–±–ª–∏—Ü—É —Å –≥–æ–¥–∞–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö.
    
    
</div>

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º.

```python
platform_years_of_life = data.pivot_table(index='platform', values='year_of_release', aggfunc=['max', 'min'])
#—Å–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –∏ —Å–æ—Ö—Ä–∞–Ω–∏–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
platform_years_of_life['life_in_years'] = platform_years_of_life['max'] - platform_years_of_life['min']
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü, –≥–¥–µ –ø–æ—Å—á–∏—Ç–∞–µ–º –≥–æ–¥–∞ –∂–∏–∑–Ω–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
platform_years_of_life['life_in_years'].mean()
#–≤—ã—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
```

–í —Ç–æ–º —á–∏—Å–ª–µ –º—ã –º–æ–≥–ª–∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ‚Äì¬†–≤—ã—á–µ—Å—Ç—å –∏–∑ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–∞—Ç—ã –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –ø–æ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –∏ –∑–∞—Ç–µ–º —É—Å—Ä–µ–¥–Ω–∏—Ç—å –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏—è.

–ü—Ä–æ–≤–µ—Ä–∏–ª –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ —Ç–µ–º –º–µ—Ç–æ–¥–æ–º, —á—Ç–æ —Ç—ã –ø—Ä–µ–¥–ª–æ–∂–∏–ª. –ù–µ–º–Ω–æ–≥–æ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–ª –≤—ã–≤–æ–¥</div>

### 3. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∞–º—ã—Ö –ø—Ä–æ–¥–∞–≤–∞–µ–º—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º.

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –≤–µ—Ä–Ω–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ –±—É–¥—É—â–µ–µ –≤–æ–∑—å–º–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö —Å 2008 –≥–æ–¥–∞ –ø–æ 2016.

```python
data_last_years = data.loc[(data.year_of_release >= 2008)] #–≤–æ–∑—å–º–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –∑–∞ 2008-2016 –≥–æ–¥–∞

(
    data_last_years
    .pivot_table(index='platform', values='sum_of_sales', aggfunc='median') #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ, –ø–æ—Å—á–∏—Ç–∞–µ–º –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º
    .sort_values(by='sum_of_sales', ascending=False) #–æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —É–±—ã–≤–∞–Ω–∏—é
    .plot(grid=True, kind='bar', sharex=False, figsize=(12,5)) #–ø–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É
)

plt.title('–°—Ä–µ–¥–Ω–∏–µ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º')
plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞', color='gray')
plt.ylabel('–°—Ä–µ–¥–Ω–∏–µ –ø—Ä–æ–¥–∞–∂–∏', color='gray')
plt.legend(['–°—Ä–µ–¥–Ω–µ–µ –ø–æ —Å—É–º–º–∞—Ä–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º'])
plt.show()
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

```python
data_last_years_platform_sales_sum_df = (
    data_last_years
    #–≤–æ–∑—å–º–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑–∞ 2008-2016 –≥–æ–¥–∞
    .pivot_table(index=['platform', 'year_of_release'], values='sum_of_sales', aggfunc='median')
    #–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º–∞—Ä–Ω–æ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ–¥—É –∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ
    .pivot_table(index='year_of_release', values='sum_of_sales', columns='platform')
    #—Ä–∞–∑–≤–µ—Ä–Ω–µ–º —Ç–∞–±–ª–∏—Ü—É
    .reset_index()
    #–æ–±–Ω—É–ª–∏–º –∏–Ω–¥–µ–∫—Å—ã
)

#—Å–æ–∑–¥–∞–¥–∏–º –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∏
fig, axs = plt.subplots(int((len(data_last_years_platform_sales_sum_df.columns)-1)/2), 2,
                        sharex='row',
                        sharey=True,
                        figsize=(15, 35))

#–∑–∞–¥–∞–¥–∏–º –æ–±—â–µ–µ –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã
fig.suptitle('–°—Ä–µ–¥–Ω–∏–∏ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∑–∞ 2008-2016 –≥–æ–¥–∞') 

#–ø–µ—Ä–µ–∑–∞–ø–∏—à–µ–º —Ä–∞–Ω–µ–µ —Å–æ–∑–¥–∞–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –≤—Å–µ—Ö –ª–µ—Ç –æ—Ç 2008 –¥–æ 2017 –≥–æ–¥–æ–≤ –∏ –∑–∞–ø–æ–ª–Ω–∏–º –µ–µ –≤ —Ü–∏–∫–ª–µ
labels_xticks = []
for i in range(10):
    labels_xticks.append(2007+i)

counter=0 #—Å–æ–∑–¥–∞–¥–∏–º —Å—á–µ—Ç—á–∏–∫
    
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –≤—Å–µ—Ö —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ —Å—É–º–º–∞–º –ø—Ä–æ–¥–∞–∂
for i, ax in enumerate(axs):
    for j in range(2):
        counter += 1 #–¥–æ–±–∞–≤–ª—è–µ–º –µ–¥–∏–Ω–∏—Ü—É –∫ —Å—á–µ—Ç—á–∏–∫—É
        axs[i][j].plot(data_last_years_platform_sales_sum_df['year_of_release'],
                       data_last_years_platform_sales_sum_df.iloc[:,counter], 'o-') #–Ω–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫–∏  
        axs[i][j].grid() #–¥–æ–±–∞–≤–∏–º —Å–µ—Ç–∫—É
        axs[i][j].set_ylabel('–ü—Ä–æ–¥–∞–∂–∏, –º–ª–Ω.', color='gray') #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å–∏ y
        axs[i][j].xaxis.set_major_locator(ticker.FixedLocator(labels_xticks)) #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å–∏ —Ö
        axs[i][j].title.set_text(data_last_years_platform_sales_sum_df.columns[counter]) #–¥–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
        axs[i][j].set_xticklabels(labels_xticks, rotation=45) #–ø–æ–≤–µ—Ä–Ω–µ–º —Ç–∏–∫–∏ –Ω–∞ 45 –≥—Ä–∞–¥—É—Å–æ–≤

        
plt.subplots_adjust(hspace=0.3) #–¥–æ–±–∞–≤–∏–º –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –º–µ–∂–¥—É –≥—Ä–∞—Ñ–∏–∫–∞–º–∏
plt.show() #–æ—Ç–æ–±—Ä–∞–∑–∏–º –≥—Ä–∞—Ñ–∏–∫–∏
#–∑–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ, –ø–æ–¥–ø–∏—Å–∏ –ø–æ –æ—Å—è–º, –ª–µ–≥–µ–Ω–¥–∞
```

–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –Ω–∞—à–∏ –¥–µ–π—Å—Ç–≤–∏—è –ø–æ–≤—Ç–æ—Ä—è—é—Ç—Å—è —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: ‚úÖ</b>

–ö—Å—Ç–∞—Ç–∏ —ç—Ç–æ –µ—â–µ –±–æ–ª–µ–µ –º–µ–Ω–µ–µ —Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º. –í –±–∏–∑–Ω–µ—Å–µ, –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ä–µ–¥–∫–æ –º–æ–∂–Ω–æ –≤—Å—Ç—Ä–µ—Ç–∏—Ç—å –ø—Ä—è–º "–∏–¥–µ–∞–ª—å–Ω–æ–µ" –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ. –ü–æ—ç—Ç–æ–º—É, –∑–¥–µ—Å—å –º—ã –º–æ–≥–ª–∏ –≤–æ–≤—Å–µ –æ–±–æ–π—Ç–∏—Å—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –¢-—Ç–µ—Å—Ç–æ–º. 
    
–°–∞–º F-—Ç–µ—Å—Ç —Å–¥–µ–ª–∞–Ω –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ.
    
</div>

–°–æ–∑–¥–∞–¥–∏–º 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ –∑–∞–ø–∏—à–µ–º —Ç—É–¥–∞ —Å—Ä–µ–∑—ã –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ –¥–ª—è PC –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∑–∞ 2008-2016 –≥–æ–¥–∞.

```python
data_xb_last_years = data_last_years.loc[(data_last_years.platform == 'xone') &
                                         (data_last_years.user_score.isna() == False)] 
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ Xbox One –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
data_pc_last_years = data_last_years.loc[(data_last_years.platform == 'pc') &
                                         (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PC –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
```

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã. –î–ª—è —ç—Ç–æ–≥–æ –ø—Ä–æ–≤–µ–¥–µ–º F-—Ç–µ—Å—Ç –º–µ—Ç–æ–¥–æ–º CDF.
  
–ì–∏–ø–æ—Ç–µ–∑—ã:
  
H0: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–≤–Ω—ã \
H1: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã.

```python
dfn = len(data_xb_last_years.user_score) - 1 #–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –¥–ª—è –ø–µ—Ä–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
dfd = len(data_pc_last_years.user_score) - 1 #–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –¥–ª—è –ø–µ—Ä–≤–æ–π –≤—ã–±–æ—Ä–∫–∏

stat = np.var(data_xb_last_years.user_score, ddof=1)/np.var(data_pc_last_years.user_score, ddof=1) #–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É

p = st.stats.f.cdf(stat, dfn, dfd) #–ø—Ä–æ–≤–µ–¥–µ–º F-—Ç–µ—Å—Ç

print(f'stat={stat:.3f}, p={p:.3f}') #–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value
if p > 0.05: #—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5% –Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1
    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

–°–¥–µ–ª–∞–ª –ø–æ–∫–æ–º–ø–∞–∫—Ç–Ω–µ–µ</div>

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤ —ç—Ç–∏—Ö –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É `user_score` —Å –ø–æ–º–æ—â—å—é Q-Q plot.  
–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞ –≤ —á–∏—Å–ª–∞—Ö Z-—à–∫–∞–ª—ã.

```python
def qq_plot(x, title): #—Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞ qq-plot
    m = x.mean() #–≤—ã—Å—á–∏—Ç–∞–µ–º –º–µ–¥–∏–∞–Ω—É –≤ –≤—ã–±–æ—Ä–∫–µ
    std = x.std() #–≤—ã—Å—á–∏—Ç–∞–µ–º –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
    x = (x - m)/std #–ø–µ—Ä–µ–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –≤ Z-—à–∫–∞–ª—É

    sm.qqplot(x, line ='45') #—Å–æ–∑–¥–∞–¥–∏–º –≥—Ä–∞—Ñ–∏–∫ Q-Q plot
    plt.title(title)
    plt.show()
```

```python
qq_plot(data_xb_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã Xbox One. 2008-2016') 
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ Xbox One
```

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: ‚úÖ</b>

–û—Ç–ª–∏—á–Ω—ã–π –∏ –∫–ª–∞—Å—Å–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö üëç
    
</div>

```python
qq_plot(data_pc_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã PC. 2008-2016')
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PC
```

–¢–∞–∫ –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ –≤ –Ω–∞—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏ –Ω–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–æ –ø—Ä–∏–º–µ–Ω–∏–º U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö.

```python
stat, p = st.stats.mannwhitneyu(data_xb_last_years.user_score, data_pc_last_years.user_score) 
#–ø—Ä–∏–º–µ–Ω–∏–º —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏

print(f'stat={stat:.3f}, p={p:.3f}') 
#–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value

if p > 0.05: 
#—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5% 
#–Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1

    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

–° –ø–æ–º–æ—â—å—é —Ç–µ—Å—Ç–∞ –º—ã –º–æ–∂–µ–º –æ—Ç–∫–∏–Ω—É—Ç—å H0 –∏ –ø—Ä–∏–º–∏–Ω–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É H1 –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã.

### 2. –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã:  
  
H0: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ `Action` –∏ `Sports` —Ä–∞–≤–Ω—ã \
H1: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ —É –∂–∞–Ω—Ä–æ–≤ `Action` –∏ `Sports` –Ω–µ—Ä–∞–≤–Ω—ã.

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: ‚úÖ</b>

–ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ.
    
    
    
</div>

–°–æ–∑–¥–∞–¥–∏–º 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ –∑–∞–ø–∏—à–µ–º —Ç—É–¥–∞ —Å—Ä–µ–∑—ã –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –∑–∞ 2008-2016 –≥–æ–¥–∞.

```python
data_action_last_years = data_last_years.loc[(data_last_years.genre == 'action') & (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –∂–∞–Ω—Ä—É action –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
data_sports_last_years = data_last_years.loc[(data_last_years.genre == 'sports') & (data_last_years.user_score.isna() == False)]
#–∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–∞–Ω–Ω—ã–µ –ø–æ –∂–∞–Ω—Ä—É sports –∏ –æ—Ç–∫–∏–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è NaN –≤ —Å—Ç–æ–ª–±—Ü–µ user_score
```

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã. –î–ª—è —ç—Ç–æ–≥–æ –ø—Ä–æ–≤–µ–¥–µ–º F-—Ç–µ—Å—Ç –º–µ—Ç–æ–¥–æ–º CDF.
  
–ì–∏–ø–æ—Ç–µ–∑—ã:
  
H0: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–≤–Ω—ã \
H1: –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã.

```python
dfn = len(data_action_last_years.user_score) - 1 #–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –¥–ª—è –ø–µ—Ä–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
dfd = len(data_sports_last_years.user_score) - 1 #–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –¥–ª—è –ø–µ—Ä–≤–æ–π –≤—ã–±–æ—Ä–∫–∏

stat = np.var(data_action_last_years.user_score, ddof=1)/np.var(data_sports_last_years.user_score, ddof=1)
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É

p = st.stats.f.cdf(stat, dfn, dfd) #–ø—Ä–æ–≤–µ–¥–µ–º F-—Ç–µ—Å—Ç

print(f'stat={stat:.3f}, p={p:.3f}') #–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value
if p > 0.05:
#—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5%
#–Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1
    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

F-—Ç–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –Ω–µ —Ä–∞–≤–Ω—ã.

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –≤ —ç—Ç–∏—Ö –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É `user_score` —Å –ø–æ–º–æ—â—å—é Q-Q plot.  

```python
qq_plot(data_action_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–∞ Action. 2008-2016') #–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä—É Action
```

```python
qq_plot(data_sports_last_years.user_score, '–ì—Ä–∞—Ñ–∏–∫ Q-Q plot –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–∞ Sports. 2008-2016') #–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ user_score —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä—É Sports
```

–¢–∞–∫ –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ –≤ –Ω–∞—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏ –Ω–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–æ –ø—Ä–∏–º–µ–Ω–∏–º U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö.

```python
stat, p = st.stats.mannwhitneyu(data_action_last_years.user_score, data_sports_last_years.user_score) #–ø—Ä–∏–º–µ–Ω–∏–º —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∏-–£–∏—Ç–Ω–∏
print(f'stat={stat:.3f}, p={p:.3f}') #–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value
if p > 0.05: #—Å–æ–∑–¥–∞–¥–∏–º —É—Å–ª–æ–≤–∏–µ, —á—Ç–æ –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ p-value –±–æ–ª—å—à–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤ 5% –Ω–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –æ—Ç–∫–ª–æ–Ω—è–µ–º H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1
    print('–ù–µ—Ç –¥–æ—Å—Ç–æ—á–Ω—ã—Ö –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0')
else:
    print('–ï—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π, —á—Ç–æ–±—ã –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0, –ø—Ä–∏–Ω–∏–º–∞–µ–º H1.')
```

–¢–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª, —á—Ç–æ –º–æ–∂–µ–º –æ—Ç–∫–ª–æ–Ω–∏—Ç—å H0 –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º H1, —Ç.–µ. —Å—Ä–µ–¥–Ω–∏–∏ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ –Ω–µ—Ä–∞–≤–Ω—ã. –°—Ç—Ä–∞–Ω–Ω–æ, —á—Ç–æ p-value –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ 0.

### 3. –í—ã–≤–æ–¥

–ò–∑ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã—Ö –≥–∏–ø–æ—Ç–µ–∑ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –Ω–∏ –æ–¥–Ω–∞ –Ω–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–ª–∞—Å—å. –í –ø–µ—Ä–≤–æ–π —á–∞—Å—Ç–∏ –º—ã –ø—Ä–æ–≤–µ—Ä—è–ª–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –≤—ã–±–æ—Ä–∫–∞—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º `Xbox One` –∏ `PC`, –∞ –≤–æ –≤—Ç–æ—Ä–æ–π —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ –≤—ã–±–æ—Ä–∫–∞–º –≤ –∂–∞–Ω—Ä–∞—Ö `Action` –∏ `Sports`.
